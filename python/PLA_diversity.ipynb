```python
import numpy as np
import pandas as pd

# 导入数据函数# 导入数据函 
def loadData(filename):
    data = pd.read_csv(filename, sep='\s+', header=None)
    data = data.as_matrix()
    col, row = data.shape
    X = np.c_[np.ones((col, 1)), data[:, 0: row-1]]
    Y = data[:, row-1:row]
    return X, Y
def W_error(X,Y,W):
    halts = 0
    row, col = X.shape
    for i in range(row):
        score = X[i:i+1,:].dot(W)
        #print('score=%s,y=%s' % (score,Y[i]))
        if(score>0.0):
            sign = 1      
        else:
            sign = -1
        if( sign != Y[i]):
            halts = halts + 1
    return halts
def pocket_algorithm(X, Y, eta):
    row, col = X.shape
    theta = np.zeros((col, 1))
    num = 0; prevpos = 0
    halts = 0
    
    for i in range(0,row):
        score = X[i:i+1,:].dot(theta)
        #print('thetatype:',theta.shape)
        if score == 0.0: score = -1
        if np.sign(score) != int(Y[i]):
            halts += 1
            theta += eta*Y[i,0]*X[i:i+1, :].T
    
    print ("halts:" + str(halts) )
    return theta
# Q15-Q17导入数据项
X, Y = loadData('data\hw1_15_train.dat')
eta = 1
theta = pocket_algorithm(X, Y, eta)
for w in theta:
    print ('w:',w)
halts = W_error(X,Y,theta)
print("error:",halts/len(X))

    
    
```
