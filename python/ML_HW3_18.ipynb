```python
import numpy as np
import pandas as pd
import math
import scipy.linalg as lin
import matplotlib.pyplot as plt
# 导入数据函数
def loadData(filename):
    data = pd.read_csv(filename, sep='\s+', header=None)
    data = data.as_matrix()
    col, row = data.shape
    X = np.c_[np.ones((col, 1)), data[:, 0: row-1]]
    Y = data[:, row-1:row]
    return X, Y
# sigmoid函数
def sigmoid(z):
    zback = 1/(1+np.exp(-1*z))
    return zback
# Logistic Regression
def logisticReg(X, Y, eta, numiter, flag=0):
    row, col = X.shape
    theta = np.zeros((col, 1))
    num = 0
    for i in range(numiter):
        if flag == 0:
            derr = (-1*X*Y).T.dot(sigmoid(-1*X.dot(theta)*Y))/row
        
        else:
            print('fuckfuckfuck')
            if num >= row:
                num = 0
            derr = -Y[num, 0]*X[num:num+1, :].T*sigmoid(-1*X[num, :].dot(theta)*Y[num, 0])
            print('derr:',X[num, :].dot(theta)[0])
            num += 1
        theta -= eta*derr
    return theta
# 误差计算函数
def mistake(X, Y, theta):
    yhat = X.dot(theta)
    yhat[yhat > 0] = 1
    yhat[yhat <= 0] = -1
    err = np.sum(yhat != Y)/len(Y)
    return err
```
