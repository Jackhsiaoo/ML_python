```python
import sys
import numpy as np
import math
from random import *
import pandas as pd
# 导入数据函数
def loadData(filename):
    data = pd.read_csv(filename, sep='\s+', header=None)
    data = data.as_matrix()     ##turn into matrix
    row, col = data.shape
    X = np.c_[np.ones((row, 1)), data[:, 0: col-1]]   #[:,0:4]=0,1,2,3
    Y = data[:, col-1:col]  #4:5>>4
    return X, Y
def W_error(X,Y,W):
    halts = 0
    row, col = X.shape
    for i in range(row):
        score = X[i:i+1,:].dot(W)
        #print('score=%s,y=%s' % (score,Y[i]))
        if(score>0.0):
            sign = 1      
        else:
            sign = -1
        if( sign != Y[i]):
            halts = halts + 1
            W += Y[i]*X[i:i+1, :].T
    return halts
def pocket_algorithm(X, Y, r):
    best_W = np.zeros((5,1))
    best_error = W_error(X, Y, best_W)
    W = np.zeros((5,1))
    rounds = 0
    while rounds<100:
        # get random one row of X
        rand=r.randint(0,len(X)-1)  
        X_rand = X[rand:rand+1,:] #len(X)= total row 
        y = Y[rand]
        score = X_rand.dot(W)
        if(score>0.0):
            sign = 1      
        else:
            sign = -1
        if( sign != y):
            rounds = rounds + 1
            W += y*X_rand.T
            # update best_W
            curr_error = W_error(X, Y, W)  #w_test
            #print ("curr_error:" + str(curr_error) + ",best_error:" + str(best_error) ) 
            if curr_error<best_error:
                best_W=W
                best_error = curr_error
    print('best_error=:'+str(best_error))
    return best_W
    
    
if __name__ == '__main__':   
    # Q17导入数据项
    X_train, Y_train = loadData('data\hw1_18_train.dat')
    X_test, Y_test = loadData('data\hw1_18_test.dat')
    # iteratively pocket algorithm
    iterative_times = 100
    total_error_times = 0
    r = Random()
    for i in range(0,iterative_times):
        # each round initialize a random seed
        r.seed(i)
        # conduct one round pocket algorithm
        W = pocket_algorithm(X_train, Y_train, r)
        # accmulate error times
        error_times = W_error(X_test, Y_test, W)
        total_error_times = total_error_times + error_times
    print (str( (1.0*total_error_times)/(iterative_times*len(X_test)) ) )
```
